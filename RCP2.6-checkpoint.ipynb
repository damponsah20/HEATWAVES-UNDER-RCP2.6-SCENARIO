```python
ds_data = xr.open_dataset('tasmax_GHA-22_MPI-M-MPI-ESM-LR_rcp26_r1i1p1_GERICS-REMO2015_v1_day_20060101-21001231.nc').sel(time=slice('2025','2054'))
ds_data
```

```python

# Convert the temperature data from Kelvin to Celsius
ds_data['tasmax'] = ds_data['tasmax'] - 273.15

# Select locations
locations = {
    'Bole': ds_data.sel(rlon=-2.53, rlat=9.13, method='nearest'),
    'Navrongo': ds_data.sel(rlon=-1.0921, rlat=10.8956, method='nearest'),
    'Wa': ds_data.sel(rlon=-2.5019, rlat=10.0607, method='nearest'),
    'Walewale': ds_data.sel(rlon=-0.8, rlat=10.35, method='nearest'),
    'Yendi': ds_data.sel(rlon=-0.0099, rlat=9.4427, method='nearest')
}
```

```python
# Calculate 95th percentile for each location and filter consecutive days
percentiles = {name: np.percentile(loc['tasmax'].values, 95) for name, loc in locations.items()}
for name, loc in locations.items():
    loc['above_95th'] = loc['tasmax'] > percentiles[name]
```

```python
# Calculate 95th percentile for each location and filter consecutive days
consecutive_days = {}

for name, loc in locations.items():
    # Calculate 95th percentile
    percentile_95 = np.percentile(loc['tasmax'].values, 95)
    
    # Identify days exceeding the 95th percentile
    loc['above_95th'] = loc['tasmax'] > percentile_95
    
    # Converting time to a more usable format
    loc['date'] = loc['time'].dt.strftime('%Y-%m-%d')
    
    # Filtering for consecutive days
    df = loc[['tasmax', 'above_95th', 'date']].to_dataframe().reset_index()
    df['consecutive'] = df['above_95th'].astype(int).groupby((df['above_95th'] != df['above_95th'].shift()).cumsum()).cumsum()
    consecutive_days[name] = df[(df['consecutive'] >= 3) & df['above_95th']]

yendi_consecutive = consecutive_days['Yendi']
yendi_consecutive
```

```python
# Defining the function to filter for consecutive days above the 95th percentile and extract episodes
def filter_consecutive_days(loc):
    df = loc[['tasmax', 'above_95th']].to_dataframe().reset_index()
    df['consecutive'] = df['above_95th'].astype(int).groupby((df['above_95th'] != df['above_95th'].shift()).cumsum()).cumsum()
    df['episode_id'] = (df['above_95th'] & (df['consecutive'] == 1)).cumsum()
    episode_starts = df[df['consecutive'] == 1].reset_index()
    episode_ends = df[df['consecutive'] == df['consecutive'].groupby(df['episode_id']).transform('max')].reset_index()
    episodes = pd.merge(episode_starts, episode_ends, on='episode_id', suffixes=('_start', '_end'))
    episodes = episodes[episodes['consecutive_end'] >= 3]
    return episodes[['time_start', 'time_end', 'tasmax_start', 'tasmax_end', 'episode_id']]

# Getting the episodes for each location
episodes = {name: filter_consecutive_days(loc) for name, loc in locations.items()}

# Defining the function to calculate the mean magnitude for each episode
def calculate_episode_magnitude(episodes, original_loc):
    magnitudes = []
    for _, episode in episodes.iterrows():
        episode_df = original_loc.sel(time=slice(episode['time_start'], episode['time_end']))
        daily_magnitudes = episode_df['tasmax'] - episode_df['tasmax'].min()
        mean_magnitude = daily_magnitudes.mean().item()
        magnitudes.append(mean_magnitude)
    episodes['mean_magnitude'] = magnitudes
    return episodes

# Calculating the mean magnitude for each episode
for name in locations.keys():
    episodes[name] = calculate_episode_magnitude(episodes[name], locations[name])

# Creating colormap
cmap = LinearSegmentedColormap.from_list("custom_cmap", ["yellow","orange", "red"])

# Defining the plotting function for heat wave episodes with vertical bars and dates on the y-axis
def plot_heat_wave_episodes(episodes_df, location_name, percentile):
    # Flattening the start and end dates into a single list of dictionaries
    episodes = [
        {
            'date_start': episode['time_start'],
            'date_end': episode['time_end'],
            'mean_magnitude': episode['mean_magnitude'],
            'duration': (episode['time_end'] - episode['time_start']).days + 1
        }
        for _, episode in episodes_df.iterrows()
    ]

    # Creating DataFrame from episodes list
    episodes_df = pd.DataFrame(episodes)

    # Sorting the episodes by the start date to ensure chronological order
    episodes_df = episodes_df.sort_values(by='date_start')

    # Creating a normalized colormap based on the mean magnitude
    norm = Normalize(vmin=episodes_df['mean_magnitude'].min(), vmax=episodes_df['mean_magnitude'].max())



   # Plotting the data as bars from start date to end date
    fig, ax = plt.subplots(figsize=(14, 10))

    for idx, episode in episodes_df.iterrows():
        color = cmap(norm(episode['mean_magnitude']))
        y_pos = episode['date_start'].timetuple().tm_yday
        ax.bar(episode['date_start'].year, episode['duration'], bottom=y_pos, color=color, width=0.6, edgecolor='black')
        ax.text(episode['date_start'].year-0.35, y_pos + episode['duration'] / 2, f'{episode["duration"]}', ha='right', va='center', color='black', fontsize=9, fontweight='bold')

    ax.annotate(f'95th Percentile: {percentile:.2f}°C', xy=(0.99, 0.99), xycoords='axes fraction', fontsize=12, fontweight='bold', ha='right', va='top',
                bbox=dict(boxstyle='round,pad=0.5', edgecolor='black', facecolor='white'))
    ax.set_xticks(np.arange(episodes_df['date_start'].dt.year.min(), episodes_df['date_start'].dt.year.max() + 1))
    ax.set_xticklabels(np.arange(episodes_df['date_start'].dt.year.min(), episodes_df['date_start'].dt.year.max() + 1), rotation=45, fontweight='bold')
    ax.set_yticks(np.arange(0, 151, 30))
    ax.set_yticklabels([f'{i}' for i in range(1, 152, 30)], fontweight='bold')
    ax.set_xlabel('YEAR', fontweight='bold', fontsize=12)
    ax.set_ylabel('DAY OF THE YEAR', fontweight='bold', fontsize=12)
    ax.set_title(f'MEAN MAGNITUDE OF HEATWAVE EPISODES OVER {location_name.upper()} USING R.C.P 2.6 (2025-2054)', fontweight='bold', fontsize=14)
    sm = ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    cbar = fig.colorbar(sm, ax=ax, label='MEAN MAGNITUDE (°C)')
    cbar.set_label('MEAN MAGNITUDE (°C)', fontsize=12, fontweight='bold')
    plt.grid(True, alpha=0.9)
    plt.tight_layout()
    plt.savefig(f'RCP_2.6_heatwave_magnitude_{location_name}.png',dpi=1000)
    plt.show()

# Plotting for each location
for location in locations.keys():
    plot_heat_wave_episodes(episodes[location], location, percentiles[location])
```

```python
# Defining the function to filter for consecutive days above the 95th percentile and extract episodes
def filter_consecutive_days(loc):
    df = loc[['tasmax', 'above_95th']].to_dataframe().reset_index()
    df['consecutive'] = (df['above_95th'] != df['above_95th'].shift()).cumsum()
    df['episode_id'] = df.groupby('consecutive').ngroup() + 1
    episode_starts = df[df['above_95th'] & (df['above_95th'].shift() == False)]
    episode_ends = df[df['above_95th'] & (df['above_95th'].shift(-1) == False)]
    
    episodes = pd.merge(episode_starts, episode_ends, on='episode_id', suffixes=('_start', '_end'))
    episodes['duration'] = (episodes['time_end'] - episodes['time_start']).dt.days + 1
    episodes = episodes[episodes['duration'] >= 3]
    
    return episodes[['time_start', 'time_end', 'tasmax_start', 'tasmax_end', 'episode_id', 'duration']]

# Defining the function to calculate the mean magnitude for each episode
def calculate_episode_magnitude(episodes, original_loc):
    magnitudes = []
    for _, episode in episodes.iterrows():
        episode_df = original_loc.sel(time=slice(episode['time_start'], episode['time_end']))
        daily_magnitudes = episode_df['tasmax'] - episode_df['tasmax'].min()
        mean_magnitude = daily_magnitudes.mean().item()
        magnitudes.append(mean_magnitude)
    episodes['mean_magnitude'] = magnitudes
    return episodes

# Calculate and print the mean number of consecutive days per year for each location
mean_consecutive_days_per_year = {}
for name, loc in locations.items():
    episode_data = filter_consecutive_days(loc)
    episode_data = calculate_episode_magnitude(episode_data, loc)
    
    total_consecutive_days = episode_data['duration'].sum()
    
    # Assuming 'time' is a datetime index in loc and determining the time span in years
    time_span_years = (loc.time.max().dt.year - loc.time.min().dt.year) + 1
    mean_consecutive_days = total_consecutive_days / time_span_years if time_span_years > 0 else 0
    
    mean_consecutive_days_per_year[name] = mean_consecutive_days
    print(f"Mean number of consecutive days per year for {name}: {mean_consecutive_days:.2f}")

# Calculate and print the overall mean number of consecutive days per year across all locations
overall_mean_consecutive_days = sum(mean_consecutive_days_per_year.values()) / len(mean_consecutive_days_per_year) if mean_consecutive_days_per_year else 0
print(f"Overall mean number of consecutive days per year across all locations: {overall_mean_consecutive_days:.2f}")
```

```python
# Creating colormap
cmap = LinearSegmentedColormap.from_list("custom_cmap", ["yellow","orange","red"])

# Defining the plotting function for heat wave episodes with vertical bars and dates on the y-axis
def plot_heat_wave_episodes(episodes_df, location_name, percentile, original_loc):
    # Flattening the start and end dates into a single list of dictionaries
    episodes = [
        {
            'date_start': episode['time_start'],
            'date_end': episode['time_end'],
            'duration': (episode['time_end'] - episode['time_start']).days + 1,
            'temperatures': original_loc.sel(time=slice(episode['time_start'], episode['time_end']))['tasmax'].values
        }
        for _, episode in episodes_df.iterrows()
    ]

    # Creating DataFrame from episodes list
    episodes_df = pd.DataFrame(episodes)

    # Sorting the episodes by the start date to ensure chronological order
    episodes_df = episodes_df.sort_values(by='date_start')

    # Flattening all temperatures to find the overall min and max
    all_temperatures = np.concatenate(episodes_df['temperatures'].values)
    vmin = all_temperatures.min()
    vmax = all_temperatures.max()

    # Creating a normalized colormap based on the temperature values
    norm = Normalize(vmin=vmin, vmax=vmax)

    # Plotting the data as bars from start date to end date
    fig, ax = plt.subplots(figsize=(14, 10))

    for idx, episode in episodes_df.iterrows():
        for i, temp in enumerate(episode['temperatures']):
            color = cmap(norm(temp))
            y_pos = (episode['date_start'] + pd.Timedelta(days=i)).timetuple().tm_yday
            ax.bar(episode['date_start'].year, 1, bottom=y_pos, color=color, width=0.6, edgecolor='black')
        
        # Annotate duration in the middle of the bar
        y_mid_pos = episode['date_start'].timetuple().tm_yday + episode['duration'] / 2
        ax.text(episode['date_start'].year-0.35, y_mid_pos, f'{episode["duration"]}', ha='right', va='center', color='black', fontsize=9, fontweight='bold')

    ax.annotate(f'95th Percentile: {percentile:.2f}°C', xy=(0.99, 0.99), xycoords='axes fraction', fontsize=12, fontweight='bold', ha='right', va='top',
                bbox=dict(boxstyle='round,pad=0.5', edgecolor='black', facecolor='white'))
    ax.set_xticks(np.arange(episodes_df['date_start'].dt.year.min(), episodes_df['date_start'].dt.year.max() + 1))
    ax.set_xticklabels(np.arange(episodes_df['date_start'].dt.year.min(), episodes_df['date_start'].dt.year.max() + 1), rotation=45, fontweight='bold')
    ax.set_yticks(np.arange(0, 151, 30))
    ax.set_yticklabels([f'{i}' for i in range(1, 152, 30)], fontweight='bold')
    ax.set_xlabel('YEAR', fontweight='bold', fontsize=12)
    ax.set_ylabel('DAY OF THE YEAR', fontweight='bold', fontsize=12)
    ax.set_title(f'HEATWAVE EPISODES OVER {location_name.upper()} USING R.C.P 2.6 (2025-2054)', fontweight='bold', fontsize=14)
    sm = ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    cbar = fig.colorbar(sm, ax=ax, label='TEMPERATURE (°C)')
    cbar.set_label('TEMPERATURE (°C)', fontsize=12, fontweight='bold')
    plt.grid(True, alpha=0.9)
    plt.tight_layout()
    plt.savefig(f'RCP2.6_INTENSITY_{location_name}.png',dpi=1000)
    plt.show()

# Plotting for each location
for location in locations.keys():
    plot_heat_wave_episodes(episodes[location], location, percentiles[location], locations[location])
```
